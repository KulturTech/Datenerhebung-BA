#Analysis of all the tweets and press releases by the VHP, RSS and BJP on the Manipur conflict

Milind Parande's Press Release (15.05.2023)

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk import pos_tag, ne_chunk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('vader_lexicon')

# Example text
text = """
Vishva Hindu Parishad (VHP) today strongly condemned the gruesome attacks by Kukis militants on Meitei Samaj and it's places of worship in Manipur. Releasing a district wise short list of selected grand temples of Maitie samaj in Manipur hills, either Burnt or destroyed, during the clashes, the national secretary general of VHP Shri Milind parande today said that we are already serving the affected people, now the Hindu society needs to come forward to rebuild these temples. He appealed for peace in the area and warned  against spreading false narrative that only the Churches had been attacked. Shri Milind parande also said that during the unfortunate clashes between the two communities, many State properties have been destroyed. Now the Hindu society needs to come forward to rebuild the damaged/destroyed temples.
VHP appealed that Peace and restraint must be excercised and the anti-national and anti-social elements must be controlled. Elements and organisations trying to disrupt peace in the region must be taken to task strictly.
"""

# Tokenization
words = word_tokenize(text)
sentences = sent_tokenize(text)

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]

# Part-of-speech tagging
pos_tags = pos_tag(filtered_words)

# Named Entity Recognition (NER)
ner_tags = ne_chunk(pos_tags)

# Frequency Distribution
freq_dist = FreqDist(filtered_words)

# Sentiment Analysis
sia = SentimentIntensityAnalyzer()
sentiment_score = sia.polarity_scores(text)

# Display results
print("Original Text:")
print(text)

print("\nTokenization:")
print(words)

print("\nSentence Tokenization:")
print(sentences)

print("\nFiltered Words (without stopwords):")
print(filtered_words)

print("\nPart-of-Speech Tags:")
print(pos_tags)

print("\nNamed Entity Recognition:")
print(ner_tags)

print("\nWord Frequency Distribution:")
print(freq_dist)

print("\nSentiment Analysis Score:")
print(sentiment_score)
```

```python
def analyze_tweet(tweet):
    # Tokenize the tweet
    words = word_tokenize(tweet)

    # Part-of-speech tagging
    pos_tags = pos_tag(words)

    # Sentiment analysis
    sia = SentimentIntensityAnalyzer()
    sentiment_score = sia.polarity_scores(tweet)

    return words, pos_tags, sentiment_score

# Example tweet
tweet_text = "Manipur, which had been peaceful for almost a decade has suddenly seen mutual discord and conflict. Who has a vested interest in trying to disturb the atmosphere ? Which are the  forces that are attempting to trigger hatred and violence by causing an incident as soon as any positive step is taken for peace."

# Analyze the tweet
words, pos_tags, sentiment_score = analyze_tweet(tweet_text)

# Print results
print("Tokenized Words:", words)
print("Part-of-Speech Tags:", pos_tags)
print("Sentiment Score:", sentiment_score)
```
